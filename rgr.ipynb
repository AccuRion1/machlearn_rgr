```python
import os

from ultralytics import YOLO
from IPython.display import Image

dataset_path = 'Clothes_Dataset'
print(os.listdir(dataset_path))
```

```python
subfolders = ['train', 'valid', 'test']
for sf in subfolders:
    print(f"\n--- {sf.upper()} ---")
    img_path = os.path.join(dataset_path, sf, "images")
    label_path = os.path.join(dataset_path, sf, "labels")

    print("Изображения:", len(os.listdir(img_path)))
    print("Разметки:", len(os.listdir(label_path)))
```

```python
model = YOLO("yolov8n.pt")

results = model.train(
    data="Clothes_Dataset/data.yaml",
    epochs=50,
    imgsz=640,
    batch=8,
    device=0
)
```
